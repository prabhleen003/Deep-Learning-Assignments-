```python
import tensorflow as tf
```

```python
from tensorflow import keras
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
X=tf.linspace(-10,10,100)
X
```

```python
X.shape
```

```python
#Sigmoid Activation Funciton
y=tf.keras.activations.sigmoid(X)

plt.plot(X,y,label="sigmoid(x)=1/(1+exp(-x))");
plt.title("sigmoid")
plt.xlabel("X")
plt.ylabel("Sigmoid of X")
plt.legend()
plt.show()
```

```python
# tanh activation function
y=tf.keras.activations.tanh(X)

plt.plot(X,y,label="tanh");
plt.title("Hyperbolic Tangent function")
plt.xlabel("X")
plt.ylabel("tanh(X)")
plt.legend()
plt.show()
```

```python
# Relu activation function
y=tf.keras.activations.relu(X)

plt.plot(X,y,label="y=max(x,0)");
plt.title("Relu activation function")
plt.xlabel("X")
plt.ylabel("max(x,0)")
plt.legend()
plt.show()
```

```python
# Leaky Relu
y=tf.nn.leaky_relu(X)

plt.plot(X,y,label="Leaky Relu");
plt.title("Leaky Relu")
plt.xlabel("X")
plt.ylabel("Leaky Relu")
plt.legend()
plt.show()
```

```python
# Elu (Exponetial Linear Unit)
y=tf.keras.activations.elu(X,alpha=0.9)

plt.plot(X,y,label="y=x if x>0 else alpha*(exp(x)-1) if x<0");
plt.title("Elu")
plt.xlabel("X")
plt.ylabel("Elu")
plt.legend()
plt.show()
```

```python
# Softmax Activation funciton
x=tf.constant([150,50,10],dtype=tf.float32)
x=tf.expand_dims(x,axis=0)
print(x.shape,x)
y=tf.keras.activations.softmax(x)
print(y.shape,y)
```
