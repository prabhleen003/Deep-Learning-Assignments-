```python
import numpy as np

class Perceptron:
  def __init__(self,learning_rate=-.01,num_iterations=1000):
    self.lr = learning_rate
    self.n_iters = num_iterations
    self.activation_function=self._relu_function
    self.weights = None
    self.bias = None

  def _relu_function(self,x):
    return np.maximum(0,x)

  def fit(self,X,y):
    n_samples, n_features = X.shape
    self.weights=np.zeros(n_features)
    self.bias=0

    for _ in range(self.n_iters):
      for idx,x_i in enumerate(X):
        linear_output=np.dot(x_i,self.weights)+self.bias
        y_predicted=self.activation_function(linear_output)

        update=self.lr*(y[idx]-y_predicted)
        self.weights+=update*x_i
        self.bias+=update

  def predict(self,X):
    linear_output=np.dot(X,self.weights)+self.bias
    y_predicted=self.activation_function(linear_output)
    return y_predicted
```

```python
X=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,0,0,1])
```

```python
perceptron=Perceptron(learning_rate=0.1,num_iterations=10)
perceptron.fit(X,y)
```

```python
perceptron.predict(X)
```
